```python
# get friends first and last name

import requests

def get_friend(user_id):
    my_request = requests.get(f'https://api.vk.com/method/friends.get?fields=bdate&user_id={user_id}&v=5.92&access_token={token}')
    full_request = my_request.json()['response']['items']
    for i in full_request:
        print(i['first_name'], i['last_name'])

get_friend(531301803)
```

```python
# getting users online status from vk and sending whatapp message

import os
from dotenv import load_dotenv
from twilio.rest import Client
import requests
import time

load_dotenv()
token = os.getenv('Token_vk')

def get_status_online(user_id):
    request_ = requests.get(f'https://api.vk.com/method/users.get?fields=online&user_id={user_id}&v=5.92&access_token={token}')
    return request_.json()['response'][0]['online']

def send_message(data):
    account_sid = os.getenv("sid_twilio")
    auth_token = os.getenv("token_test_twilio")
    
    DATA = {
        'sms_phone_twilio': '+19207173408',
        'whatsapp_phone_twilio': 'whatsapp:+14155238886',
        'reciever_phone': 'whatsapp:+79146942294',
        }

    client = Client(account_sid, auth_token)
    message = client.messages.create(
        to=DATA['reciever_phone'],
        from_=DATA['whatsapp_phone_twilio'],
        body=data)


status = None
new_status = None

while status != 2:
    time.sleep(5)
    new_status = get_status_online(713836507)  
    
    if new_status:
        body_text = "Пользователь онлайн"
    else:
        body_text = "Пользователь офлайн"
    
    if new_status != status:
        send_message(body_text)
    
    status = new_status

```

```python
from dotenv import load_dotenv
import requests
import os
from hashlib import sha256
import json

load_dotenv()
api_key_ = os.getenv('mango_api_key')
api_salt_ = os.getenv('mango_api_salt')


json_data = {
    'date_from': 1659880800,
    'date_to': 1659967199,
    #'fields': 'from_number',
    # 'call_party': {
    #     'extension': ('103', '104')
    # }
}

sign = sha256((api_key_ + json.dumps(json_data) + api_salt_).encode("UTF-8")).hexdigest()

data = {
    'vpbx_api_key': api_key_,
    'sign': sign,
    'json': json.dumps(json_data)

}
request_mango = requests.post(f'https://app.mango-office.ru/vpbx/stats/request', data=data)
key = request_mango.json()['key']
print(key)


```

```python
json_data = {
    'key': key
}

sign = sha256((api_key_ + json.dumps(json_data) + api_salt_).encode("UTF-8")).hexdigest()

data = {
    'vpbx_api_key': api_key_,
    'sign': sign,
    'json': json.dumps(json_data)

}
request_mango_stats = requests.post(f'https://app.mango-office.ru/vpbx/stats/result', data=data)
catch = request_mango_stats.text
```
